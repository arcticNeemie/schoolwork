<h2>The Basics</h2>

Some sample code for reading and displaying an image

```python
img = plt.imread("Lab 1/cameraman.png")

#Requires a list of images (if only 1, can just be an image) and nrows, ncols are integers
def displayImage(images, nrows = 1, ncols=1, title=[]):
    #Handle the case of 1 image
    if nrows == 1 and ncols == 1:
        images = [images]
    #Mismatch
    if len(images) != nrows*ncols:
        print("Number of images != number of subplots")
        return
    #Deal for various types
    type = images[0].dtype
    if np.issubdtype(type, np.floating):
        im_max = 1
    elif np.issubdtype(type, np.integer):
        im_max = np.iinfo(type).max
    fig = plt.figure()
    ax = []
    for i in range(1, ncols*nrows +1):
        image = images[i-1]
        ax.append( fig.add_subplot(nrows, ncols,i))
        if len(title)!=0:
            ax[-1].set_title(title[i-1])
        plt.axis("off")
        plt.imshow(image,vmin=0,vmax=im_max)
    plt.show()
    
displayImage(img)
```

Here we can get some properties of the image

```python
print("Image size: %d x %d" % (img.shape[0],img.shape[1]))
print("Image datatype: %s\n" % (img.dtype))
print("Maximum pixel value: %f" % (img.max()))
print("Minimum pixel value: %f" % (img.min()))
print("Mean pixel value: %f" % (img.mean()))
```

Using numpy, we can play around with the image

```python
img1 = np.flipud(img) #Flip upside-down
img2 = np.rot90(img)  #Rotate by 90*
displayImage([img,img1,img2],1,3,title=["Original","Upside-down","Rotated by 90 degrees"])
```
