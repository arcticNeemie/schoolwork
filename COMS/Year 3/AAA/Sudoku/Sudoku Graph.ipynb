```python
#Imports
import pandas as pd    #Data Analysis
import matplotlib.pyplot as plt    #Plotting
import numpy as np    #Matrices
import seaborn as sns    #Visualisation
import scipy as sp   #Science!
%matplotlib inline
```

```python
Exp1 = pd.read_csv('experiment1.csv', encoding = "ISO-8859-1")
Exp1.info()
```

```python
fig = plt.figure(figsize=(20,20))
ax = sns.pointplot(x='input',y='time',data=Exp1,capsize=0.3)
ax.set(xlabel='k',ylabel='Time (nanoseconds)')
```

```python
Exp1['time'].describe()
```

<h2>Curve Fitting</h2>

```python
X = Exp2['input'].values
Y = Exp2['time'].values
sp.optimize.curve_fit(lambda t,a,b,c,d,e: a*(b**(c*t+d))+e,  X,  Y)
```
