```python
#Imports
import pandas as pd    #Data Analysis
import matplotlib.pyplot as plt    #Plotting
import numpy as np    #Matrices
import seaborn as sns    #Visualisation
%matplotlib inline

Data = pd.read_csv('experiment1.csv', encoding = "ISO-8859-1")

Data.info()
```

```python
fig = plt.figure(figsize=(20,20))
ax = sns.pointplot(x='input',y='time',data=Data,capsize=0.3)
ax.set(xlabel='k',ylabel='Time (nanoseconds)')
```

```python
def nineToX(row):
    return 9**row['input']

DataComparison = Data
b = DataComparison.apply (nineToX,axis=1)
print(b)
```
